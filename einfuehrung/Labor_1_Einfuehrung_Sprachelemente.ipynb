```julia
# Fibonacci-Folge
# 0, 1, 1, 2, 3, 5, 8, 13
def fib(_i: int):
    if _i == 0:
        return 0
    elif _i == 1:
        return 1
    else:
        return fib(_i - 1) + fib(_i - 2)

print(fib(6))
```

- __OOP__
    - Konstruktor
    - Methoden
    - Scope
    - __Aufgaben__:
        - Definieren Sie eine Klasse mit einem Konstruktur, zwei Methoden und drei Objekt-Eigenschaften
        - Instantiieren Sie ein Objekt der obigen Klasse, wenden Sie die Methoden an, geben Sie die Eigenschaften aus

```julia
class OhMy:
    height: float
    weight: float
    income: float

    def __init__(self, h_m=1.75, w_kg=75.0, money_euro=40_000):
        self.height=h_m
        self.weight=w_kg
        self.income=money_euro

    def calc_bmi(self) -> float:
        return self.weight /(self.height**2)

    def is_wealthy(self) -> bool:
        return self.income > 69_000

    def is_healthy(self) -> bool:
        return 18.5 <= self.calc_bmi() <= 24.9
```

```julia
my = OhMy(1.8, 100, 50_000)
my.height, my.income, my.weight, my.calc_bmi(), my.is_healthy(), my.is_wealthy()
```
